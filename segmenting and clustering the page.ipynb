# reading data from html


```python
table = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M', header=0)

# Create the initial dataframe from the table
df = pd.DataFrame(data = table[0])

# Print the shape
print('The shape of the Raw Inital Datafram is: ', df.shape)

# Output the Head of the Table
df.head()
```

# remove null or not assigned in table

```python
df1=df[(df.Borough != 'Not assigned') & (df.Neighborhood != 'Not assigned')]
```

```python

df1.reset_index(drop=True, inplace=True)
```

```python
df1.rename(columns={'Postal code':'PostalCode'},inplace=True)
part_01 = pd.DataFrame(df1.groupby(
    ['PostalCode', 'Borough'])['Neighborhood'].apply(
    lambda x: ', '.join(x))).reset_index()
```
